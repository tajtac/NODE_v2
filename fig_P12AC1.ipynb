```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import palettable
import pickle
from NODE_fns import NODE_sigma_vmap
```

```python
#Load the NN
with open('savednet/' + 'P12AC1_bsxsy.npy', 'rb') as f:
    params = pickle.load(f)

#Load data
with open('training_data/' + 'P12AC1_xybsxsy.npy', 'rb') as f:
    lamb, sigma_gt = np.load(f,allow_pickle=True)

n_offx = 72
n_offy = 76
n_equi = 81
n_strx = 101
n_stry = 72
n_hf = n_offx + n_offy + n_equi + n_strx + n_stry
i_offx = n_offx
i_offy = i_offx + n_offy
i_equi = i_offy + n_equi
i_strx = i_equi + n_strx
i_stry = i_strx + n_stry
n_data = lamb.shape[0]
#Predict stresses using the NN
F = np.zeros((lamb.shape[0], 3, 3))
F[:,0,0] = lamb[:,0]
F[:,1,1] = lamb[:,1]
F[:,2,2] = 1/(lamb[:,0]*lamb[:,1])
sigma_pred = NODE_sigma_vmap(F, params)
```

```python
error = np.zeros(sigma_gt.shape[0])
rel_error = np.zeros(sigma_gt.shape[0])
dummy = np.zeros(sigma_gt.shape[0])
for i in range(sigma_gt.shape[0]):
    for j in range(2):
        error[i]+= (sigma_gt[i,j] - sigma_pred[i,j,j])**2
        dummy[i]+= sigma_gt[i,j]**2 
    error[i] = np.sqrt(error[i])
    dummy[i] = np.sqrt(dummy[i])
    rel_error[i] = error[i]/dummy[i]
offx_error = np.sum(error[:i_offx])/n_offx
offy_error = np.sum(error[i_offx:i_offy])/n_offy
equi_error = np.sum(error[i_offy:i_equi])/n_equi
strx_error = np.sum(error[i_equi:i_strx])/n_strx
stry_error = np.sum(error[i_strx:i_stry])/n_stry
synt_error = np.sum(error[n_hf:])/(n_data-n_hf)

offx_rel_error = np.sum(rel_error[:i_offx])/n_offx
offy_rel_error = np.sum(rel_error[i_offx:i_offy])/n_offy
equi_rel_error = np.sum(rel_error[i_offy:i_equi])/n_equi
strx_rel_error = np.sum(rel_error[i_equi:i_strx])/n_strx
stry_rel_error = np.sum(rel_error[i_strx:i_stry])/n_stry
synt_rel_error = np.sum(rel_error[n_hf:])/(n_data-n_hf)
```

```python
def commonformat(ax, i1, i2, lambda_1):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.plot(lamb[i1:i2  ,lambda_1],   sigma_gt[i1:i2  ,0],'.')
    ax.plot(lamb[i1:i2  ,lambda_1],   sigma_gt[i1:i2  ,1],'.')
    ax.plot(lamb[i1:i2  ,lambda_1], sigma_pred[i1:i2  ,0,0],'-', linewidth=3)
    ax.plot(lamb[i1:i2  ,lambda_1], sigma_pred[i1:i2  ,1,1],'-', linewidth=3)
    
plt.rcParams.update({'font.size': 12})
plt.rcParams["mathtext.fontset"] = 'dejavuserif'
plt.rcParams.update({'font.sans-serif':'Arial'})
plt.rcParams['axes.prop_cycle'] = plt.cycler('color', plt.cm.Blues(np.linspace(0.5, 0.9, 2)))
fig = plt.figure(figsize=(16,12))
fig.subplots_adjust(top=0.87, hspace=0.5)

ax3 = fig.add_subplot(231)
i1 = n_offx + n_offy
i2 = n_offx + n_offy + n_equi
ax3.plot(lamb[i1:i2  ,1],   sigma_gt[i1:i2  ,0],'.', label = '$\sigma_{xx}$ (GT)')
ax3.plot(lamb[i1:i2  ,1],   sigma_gt[i1:i2  ,1],'.', label = '$\sigma_{yy}$ (GT)')
ax3.plot(lamb[i1:i2  ,1], sigma_pred[i1:i2  ,0,0],'-', label = '$\sigma_{xx}$ (Pred.)', linewidth=3)
ax3.plot(lamb[i1:i2  ,1], sigma_pred[i1:i2  ,1,1],'-', label = '$\sigma_{yy}$ (Pred.)', linewidth=3)
#title = 'Equibiaxial \nAvg abs error ($E_{abs}$) = %.3f \nAvg rel error ($E_{rel}$) = %.3f' %(equi_error, equi_rel_error)
title = 'Equibiaxial \nAvg abs error ($E_{abs}$) = %.2e [MPa]' %(equi_error)
# title = 'Equibiaxial \n$E_{abs}$ = %.3f \n$E_{rel}$ = %.3f' %(equi_error, equi_rel_error)
ax3.set(xlabel = '$\lambda_{yy}$', ylabel = '$\sigma$ [MPa]', title=title)
ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.legend(loc='center left')
plt.subplots_adjust(wspace=0.3)

ax4 = fig.add_subplot(232)
i1 = n_offx + n_offy + n_equi
i2 = n_offx + n_offy + n_equi + n_strx
commonformat(ax4, i1, i2, 0)
#title = 'Str-x \n$E_{abs}$ = %.3f \n$E_{rel}$ = %.3f' %(strx_error, strx_rel_error)
title = 'Strip-x \n$E_{abs}$ = %.2e' %(strx_error)
ax4.set(xlabel = '$\lambda_{xx}$', title=title)

ax5 = fig.add_subplot(233)
i1 = n_offx + n_offy + n_equi + n_strx
i2 = n_offx + n_offy + n_equi + n_strx + n_stry
commonformat(ax5, i1, i2, 1)
#title = 'Str-y \n$E_{abs}$ = %.3f \n$E_{rel}$ = %.3f' %(stry_error, stry_rel_error)
title = 'Strip-y \n$E_{abs}$ = %.2e' %(stry_error)
ax5.set(xlabel = '$\lambda_{yy}$', title=title)

ax1 = fig.add_subplot(235)
i1 = 0
i2 = n_offx
commonformat(ax1, i1, i2, 1)
#title = 'Off-x \n$E_{abs}$ = %.3f \n$E_{rel}$ = %.3f' %(offx_error, offx_rel_error)
title = 'Off-x \n$E_{abs}$ = %.2e' %(offx_error)
ax1.set(xlabel = '$\lambda_{yy}$', ylabel = '$\sigma$ [MPa]', title=title)

ax2 = fig.add_subplot(236)
i1 = n_offx
i2 = n_offx + n_offy
commonformat(ax2, i1, i2, 1)
#title = 'Off-y \n$E_{abs}$ = %.3f \n$E_{rel}$ = %.3f' %(offy_error, offy_rel_error)
title = 'Off-y \n$E_{abs}$ = %.2e' %(offy_error)
ax2.set(xlabel = '$\lambda_{yy}$', title=title)

fig.savefig('Figures/rfig_P12AC1.jpg', dpi=400, bbox_inches='tight')
```
