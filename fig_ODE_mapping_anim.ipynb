```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
import pickle
from NODE_fns import forward_pass
matplotlib.rcParams.update({'font.size': 15})
```

# ODE Mapping Animation

```python
#Load the NN
model_name = 'P12AC1_xybsxsy'
with open('savednet/' + model_name + '.npy', 'rb') as f:
    params = pickle.load(f)
Psi1_params, Psi2_params, Psi4v_params, Psi4w_params, _, _, _, _, _, _, _, _ = params
```

```python
ndata = 24
ncurve = 11
dt = 1/ndata
hs1 = np.zeros((ncurve,ndata))
hs2 = np.zeros_like(hs1)
hs4v = np.zeros_like(hs1)
hs4w = np.zeros_like(hs1)
for j in range(ncurve):
    I_hat = j/10.0
    hs1[j,0] = I_hat
    hs2[j,0] = I_hat
    hs4v[j,0] = I_hat
    hs4w[j,0] = I_hat
    for i in range(1, ndata):
        dydt = forward_pass(np.array([hs1[j,i-1]]), Psi1_params) #slope
        hs1[j,i] = hs1[j,i-1] + dydt*dt
        dydt = forward_pass(np.array([hs4v[j,i-1]]), Psi4v_params)
        hs4v[j,i] = hs4v[j,i-1] + dydt*dt

t = np.linspace(0,1,ndata)
```

```python
plt.rcParams["mathtext.fontset"] = 'dejavuserif'
plt.rcParams.update({'font.sans-serif':'Arial'})
line1 = [[0,0], [0,1.3]]
line2 = [[1,1], [0,1.3]]
fig, ax = plt.subplots(1,2,figsize=[16*2/3,9*2/3])
fig.subplots_adjust(left=0.1, right=0.95, wspace=0.4)
ax3 = ax[0].twinx()
ax3.spines['right'].set_visible(False)
ax3.spines['left'].set_visible(False)
ax3.spines['top'].set_visible(False)
ax3.spines['bottom'].set_visible(False)
ax3.set(ylabel = '$\partial \Psi / \partial I_{4v}$', yticks=[])
ax3.yaxis.set_ticks_position('none') 
def init():
    ax[0].plot(line1[0], line1[1], linestyle='dashed', color='k', dashes=(20,20), linewidth=1)
    ax[0].plot(line2[0], line2[1], linestyle='dashed', color='k', dashes=(20,20), linewidth=1)
    ax[0].spines['right'].set_visible(False)
    ax[0].spines['left'].set_visible(False)
    ax[0].spines['top'].set_visible(False)
    ax[0].spines['bottom'].set_visible(False)
    ax[0].set(xlabel = 'Pseudo-time', ylabel = '$I_{4v}-1$', xlim=(-0.005,1.005), ylim=(-0.005,1.25))
    ax[0].yaxis.set_ticks_position('none') 
    #ax[0].text(1.03,0.6,'$\partial \Psi / \partial I_{4v}$', fontsize=18)
    
    
    ax[1].spines['right'].set_visible(False)
    ax[1].spines['top'].set_visible(False)
    ax[1].set(xlabel='$I_{4v} - 1$', ylabel='$\partial \Psi / \partial I_{4v}$', xlim=[0,1.05], ylim=[0,0.8])
    return fig,

def animate(i):
    j = i//ndata #Which curve?
    k = i%ndata  #Where on the last curve?
    if i < ndata*ncurve:
        for l in range(j-1):
            ax[0].plot(t, hs4v[l,:], c = 'k', linewidth=2)
        ax[0].plot(t[:k+1], hs4v[j,:k+1], c = 'k', linewidth=2)

        ax[1].plot(hs4v[:j,0], hs4v[:j,-1], 'k-',marker = 'o', linewidth=2)
    else:
        for l in range(ncurve-1):
            ax[0].plot(t, hs4v[l,:], c = 'k', linewidth=2)
        ax[1].plot(hs4v[:,0], hs4v[:,-1], 'k-', marker = 'o', linewidth=2)
        pass
    return fig,

def animate_step1(i):
    if i > 2:
        #ax[0].text(0.1, 0.9,'Invariants are initial \n conditions for the \n neural ODE')
        ax[0].annotate('Invariants are initial conditions \nfor the neural ODE', (0,0.7), xytext=(0.15,1.0), arrowprops={'arrowstyle':'->'})
    return fig,

def animate_step2(i):
    ax[0].annotate('The neural ODE maps the invariants \nthrough pseudo-time to the value of \nthe strain energy derivatives', 
                   (0,0.6), xytext=(0.05,1.0))
    j = i//ndata #Which curve?
    k = i%ndata  #Where on the last curve?
    for l in range(j-1):
        ax[0].plot(t, hs4v[l,:], c = 'k', linewidth=2)
    ax[0].plot(t[:k+1], hs4v[j,:k+1], c = 'k', linewidth=2)

    ax[1].plot(hs4v[:j,0], hs4v[:j,-1], 'k-', marker = 'o', linewidth=2)
    return fig,

def animate_step3(i):
    i = i + 55
    ax[0].annotate('The neural network is not a direct \nmap between inputs and outputs', 
                   (0,0.6), xytext=(0.1,1.0))
    j = i//ndata #Which curve?
    k = i%ndata  #Where on the last curve?
    for l in range(j):
        ax[0].plot(t, hs4v[l,:], c = 'k', linewidth=2)
    ax[0].plot(t[:k+1], hs4v[j,:k+1], c = 'k', linewidth=2)

    ax[1].plot(hs4v[:j,0], hs4v[:j,-1], 'k-', marker = 'o', linewidth=2)
    return fig,

def animate_step4(i):
    i = i + 55 + 18
    ax[0].annotate('Rather, the neural network is the \nright hand side of an ODE', (0,0.6), xytext=(0.1,1.0))
    j = i//ndata #Which curve?
    k = i%ndata  #Where on the last curve?
    for l in range(j):
        ax[0].plot(t, hs4v[l,:], c = 'k', linewidth=2)
    ax[0].plot(t[:k], hs4v[j,:k], c = 'k', linewidth=2)

    ax[1].plot(hs4v[:j,0], hs4v[:j,-1], 'k-', marker = 'o', linewidth=2)
    
    m = i-55-18
    mymap = [0, 4, 5]
    if m in mymap:
        ax[0].plot(t[k-1], hs4v[j,k-1], c='k', marker='o')
    return fig,

def animate_step5(i):
    i = i + 55 + 18 + 7
    ax[0].annotate('The neural network is evaluated \nat every time step to produce \nthe trajectory', (0,0.6), xytext=(0.1,1.0))
    j = i//ndata #Which curve?
    k = i%ndata  #Where on the last curve?
    for l in range(j):
        ax[0].plot(t, hs4v[l,:], c = 'k', linewidth=2)
    ax[0].plot(t[:k], hs4v[j,:k], c = 'k', linewidth=2)

    ax[1].plot(hs4v[:j,0], hs4v[:j,-1], 'k-', marker = 'o', linewidth=2)
    
    m = i-55-18 - 7
    mymap = [0, 6]
    if m in mymap:
        ax[0].plot(t[k-1], hs4v[j,k-1], c='k', marker='o')
    
    #plot the previous markers too:
    ax[0].plot(t[0], hs4v[j,0], c='k', marker='o')
    ax[0].plot(t[4], hs4v[j,4], c='k', marker='o')
    ax[0].plot(t[5], hs4v[j,5], c='k', marker='o')
    return fig,

def animate_step6(i):
    i = i + 55 + 18 + 7 + 13
    j = i//ndata #Which curve?
    k = i%ndata  #Where on the last curve?
    for l in range(j):
        ax[0].plot(t, hs4v[l,:], c = 'k', linewidth=2)
    ax[0].plot(t[:k], hs4v[j,:k], c = 'k', linewidth=2)
    ax[1].plot(hs4v[:j,0], hs4v[:j,-1], 'k-', marker = 'o', linewidth=2)
    return fig,

def animate_step7(i):
    i = ndata*ncurve
    j = i//ndata #Which curve?
    k = i%ndata  #Where on the last curve?
    for l in range(j):
        ax[0].plot(t, hs4v[l,:], c = 'k', linewidth=2)
    ax[1].plot(hs4v[:j,0], hs4v[:j,-1], 'k-', marker = 'o', linewidth=2)
    ax[1].annotate('Because trajectories of the ODE \ndo not intersect, the resulting \ninput-output map is \nnon-decreasing', 
                   (0,0.6), xytext=(0.05,0.6))
    return fig,

#don't use this. Just left for reference.
# pauseattheend = 50
# anim = animation.FuncAnimation(fig, animate, init_func=init, frames=ndata*ncurve + pauseattheend, blit=True)
# anim.save('Figures/fig_ODE_mapping_anim/ODE_mapping_animation.gif', fps=60, dpi=300)


anim = animation.FuncAnimation(fig, animate_step1, init_func=init, frames=10, blit=True)
anim.save('Figures/fig_ODE_mapping_anim/anim_step1.gif', fps=60, dpi=300)

anim = animation.FuncAnimation(fig, animate_step2, init_func=init, frames=55, blit=True)
anim.save('Figures/fig_ODE_mapping_anim/anim_step2.gif', fps=60, dpi=300)

anim = animation.FuncAnimation(fig, animate_step3, init_func=init, frames=18, blit=True)
anim.save('Figures/fig_ODE_mapping_anim/anim_step3.gif', fps=60, dpi=300)

anim = animation.FuncAnimation(fig, animate_step4, init_func=init, frames=7, blit=True)
anim.save('Figures/fig_ODE_mapping_anim/anim_step4.gif', fps=60, dpi=300)

anim = animation.FuncAnimation(fig, animate_step5, init_func=init, frames=13, blit=True)
anim.save('Figures/fig_ODE_mapping_anim/anim_step5.gif', fps=60, dpi=300)

anim = animation.FuncAnimation(fig, animate_step6, init_func=init, frames=ndata*ncurve-55-18-7-13, blit=True)
anim.save('Figures/fig_ODE_mapping_anim/anim_step6.gif', fps=60, dpi=300)

anim = animation.FuncAnimation(fig, animate_step7, init_func=init, frames=50, blit=True)
anim.save('Figures/fig_ODE_mapping_anim/anim_step7.gif', fps=60, dpi=300)
```
