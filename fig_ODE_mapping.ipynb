```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 30})
import pickle
from NODE_fns import forward_pass
```

```python
#Load the NN
model_name = 'P12AC1_xybsxsy'
with open('savednet/' + model_name + '.npy', 'rb') as f:
    params = pickle.load(f)
Psi1_params, Psi2_params, Psi4v_params, Psi4w_params, _, _, _, _, _, _, _, _ = params

# I1
ndata = 20
ncurve = 11
dt = 1/ndata
hs1 = np.zeros((ncurve,ndata))
hs2 = np.zeros_like(hs1)
hs4v = np.zeros_like(hs1)
hs4w = np.zeros_like(hs1)
for j in range(ncurve):
    I_hat = j/10.0
    hs1[j,0] = I_hat
    hs2[j,0] = I_hat
    hs4v[j,0] = I_hat
    hs4w[j,0] = I_hat
    for i in range(1, ndata):
        dydt = forward_pass(np.array([hs1[j,i-1]]), Psi1_params) #slope
        hs1[j,i] = hs1[j,i-1] + dydt*dt
        dydt = forward_pass(np.array([hs2[j,i-1]]), Psi2_params)
        hs2[j,i] = hs2[j,i-1] + dydt*dt
        dydt = forward_pass(np.array([hs4v[j,i-1]]), Psi4v_params)
        hs4v[j,i] = hs4v[j,i-1] + dydt*dt
        dydt = forward_pass(np.array([hs4w[j,i-1]]), Psi4w_params)
        hs4w[j,i] = hs4w[j,i-1] + dydt*dt
```

```python
def commonformat(ax):
    ax.plot(line1[0], line1[1], linestyle='dashed', color='k', dashes=(20,20), linewidth=0.5)
    ax.plot(line2[0], line2[1], linestyle='dashed', color='k', dashes=(20,20), linewidth=0.5)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set(xlabel = 't', xlim=(0,1.01), ylim=(-0.05,1.25))#, title = '$H$')
    ax.yaxis.set_ticks_position('none') 
line1 = [[0,0], [0,1.3]]
line2 = [[1,1], [0,1.3]]
matplotlib.rcParams.update({'font.size': 15})
plt.rcParams["mathtext.fontset"] = 'dejavuserif'
plt.rcParams.update({'font.sans-serif':'Arial'})
#plt.rcParams['font.family'] = 'serif'
plt.rcParams['axes.prop_cycle'] = plt.cycler('color', plt.cm.Blues(np.linspace(0.4, 1, ncurve)))

fig = plt.figure(figsize=(18,12))
fig.subplots_adjust(wspace=0.6, hspace=0.2)
ax = fig.add_subplot(231)
commonformat(ax)
ax.text(-0.25,0.6,'$J_1$', fontsize=18)
ax.text(1.03,0.6,'$\partial \Psi / \partial I_1$', fontsize=18)
t = np.linspace(0,1,ndata)
for i in range(ncurve):
    ax.plot(t, hs1[i,:])

ax = fig.add_subplot(232)
commonformat(ax)
ax.text(-0.25,0.6,'$J_2$', fontsize=18)
ax.text(1.03,0.6,'$\partial \Psi / \partial I_2$', fontsize=18)
t = np.linspace(0,1,ndata)
for i in range(ncurve):
    ax.plot(t, hs2[i,:])
    
ax = fig.add_subplot(234)
commonformat(ax)
ax.text(-0.25,0.6,'$J_{4v}$', fontsize=18)
ax.text(1.03,0.6,'$\partial \Psi / \partial I_{4v}$', fontsize=18)
t = np.linspace(0,1,ndata)
for i in range(ncurve):
    ax.plot(t, hs4v[i,:])
    
ax = fig.add_subplot(235)
commonformat(ax)
ax.text(-0.25,0.6,'$J_{4w}$', fontsize=18)
ax.text(1.03,0.6,'$\partial \Psi / \partial I_{4w}$', fontsize=18)
t = np.linspace(0,1,ndata)
for i in range(ncurve):
    ax.plot(t, hs4w[i,:])

ax = fig.add_subplot(233)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.plot(hs1[:,0], hs1[:,-1], c=plt.cm.Blues(0.9), linestyle='-', label='$\partial \Psi / \partial I_1$')
ax.plot(hs2[:,0], hs2[:,-1], c=plt.cm.Blues(0.9), linestyle='--', label='$\partial \Psi / \partial I_2$')
ax.plot(hs4v[:,0], hs4v[:,-1], c=plt.cm.Blues(0.9), linestyle='-.', label='$\partial \Psi / \partial I_{4v}$')
ax.plot(hs4v[:,0], hs4w[:,-1], c=plt.cm.Blues(0.9), linestyle=(0, (5, 10)), label='$\partial \Psi / \partial I_{4w}$')
ax.set(xlabel='$J_i$', ylabel='$\partial \Psi / \partial I_i$ [MJ]')
ax.legend()
fig.savefig('Figures/rfig_ODE_mapping.jpg', dpi=400, bbox_inches='tight')
pass
```
