```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import plot
matplotlib.rcParams.update({'font.size': 18})
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from material_models import GOH
```

```python
def matrix_from_stretch(lm1, lm2, lm3):
    zeros = np.zeros(lm1.shape)
    F = np.array([[lm1, zeros, zeros], [zeros, lm2, zeros], [zeros, zeros, lm3]]).transpose((2,0,1))
    C = np.einsum('...ij,...jk->...ik', F, F.transpose(0,2,1))
    return F, C
```

```python
lm1 = np.random.rand((1000))*0.8 + 0.7
lm2 = np.random.rand((1000))*0.8 + 0.7
lm3 = 1/(lm1*lm2)
zeros = np.zeros(1000)

#Special cases:
lm1_uni = np.linspace(1,1.5,20)
lm2_uni = 1/np.sqrt(lm1_uni)
lm3_uni = lm2_uni
lm1_uni2 = lm2_uni
lm2_uni2 = lm1_uni
lm3_uni2 = lm2_uni
lm1_uni = np.hstack((np.flip(lm1_uni,axis=0), lm1_uni2))
lm2_uni = np.hstack((np.flip(lm2_uni,axis=0), lm2_uni2))
lm3_uni = np.hstack((np.flip(lm3_uni,axis=0), lm3_uni2))

lm1_equi = np.linspace(1,1.5,20)
lm2_equi = lm1_equi
lm3_equi = 1/(lm1_equi*lm2_equi)
zeros_equi = np.zeros(lm1_equi.shape[0])

lm1_strx = np.linspace(1,1.5,20)
lm2_strx = np.ones(lm1_strx.shape[0])
lm3_strx = 1/(lm1_strx*lm2_strx)
lm1_stry = lm2_strx
lm2_stry = lm1_strx
lm3_stry = lm3_strx
lm1_str = np.hstack((np.flip(lm1_strx,axis=0), lm1_stry))
lm2_str = np.hstack((np.flip(lm2_strx,axis=0), lm2_stry))
lm3_str = np.hstack((np.flip(lm3_strx,axis=0), lm3_stry))

lm1_offx = np.linspace(1,1.5,20)
lm2_offx = np.sqrt(lm1_offx)
lm3_offx = 1/(lm1_offx*lm2_offx)
lm1_offy = lm2_offx
lm2_offy = lm1_offx
lm3_offy = lm3_offx
lm1_off = np.hstack((np.flip(lm1_offx,axis=0), lm1_offy))
lm2_off = np.hstack((np.flip(lm2_offx,axis=0), lm2_offy))
lm3_off = np.hstack((np.flip(lm3_offx,axis=0), lm3_offy))
```

```python
F, C = matrix_from_stretch(lm1, lm2, lm3)
# divide the random scatter points to those above and below the uniaxial lines (feasible and infeasible)
feasibility = np.array(C[:,0,0], dtype=bool)
for i in range(C.shape[0]):
    if F[i,0,0] > 1:
        if F[i,0,0]*F[i,1,1]**2 > 1:
            feasibility[i] = True
        else:
            feasibility[i] = False
    elif F[i,1,1] > 1:
        if F[i,1,1]*F[i,0,0]**2 > 1:
            feasibility[i] = True
        else:
            feasibility[i] = False
        pass
    else:
        feasibility[i] = False
lm1_feasible = lm1[feasibility]
lm2_feasible = lm2[feasibility]
lm1_infeasible = lm1[~feasibility]
lm2_infeasible = lm2[~feasibility]
C_feasible = C[feasibility]
C_infeasible = C[~feasibility]
#Special cases:
F_uni, C_uni = matrix_from_stretch(lm1_uni, lm2_uni, lm3_uni)
F_equi, C_equi = matrix_from_stretch(lm1_equi, lm2_equi, lm3_equi)
F_str, C_str = matrix_from_stretch(lm1_str, lm2_str, lm3_str)
F_off, C_off = matrix_from_stretch(lm1_off, lm2_off, lm3_off)
```

```python
I1 = np.trace(C, axis1 = 1, axis2 = 2)
I2 = 0.5*(I1**2 - np.trace(C**2, axis1 = 1, axis2 = 2))
v0 = np.array([1.0,0.0,0.0])
w0 = np.array([0.0,1.0,0.0])
V0 = np.outer(v0, v0)
W0 = np.outer(w0, w0)
I4v = np.einsum('pij,jk->p', C, V0)
I4w = np.einsum('pij,jk->p', C, W0)

I1_feasible = np.trace(C_feasible, axis1 = 1, axis2 = 2)
I2_feasible = 0.5*(I1_feasible**2 - np.trace(C_feasible**2, axis1 = 1, axis2 = 2))
I4v_feasible = np.einsum('pij,jk->p', C_feasible, V0)
I4w_feasible = np.einsum('pij,jk->p', C_feasible, W0)
I1_infeasible = np.trace(C_infeasible, axis1 = 1, axis2 = 2)
I2_infeasible = 0.5*(I1_infeasible**2 - np.trace(C_infeasible**2, axis1 = 1, axis2 = 2))
I4v_infeasible = np.einsum('pij,jk->p', C_infeasible, V0)
I4w_infeasible = np.einsum('pij,jk->p', C_infeasible, W0)

#Special cases:
I1_uni = np.trace(C_uni, axis1 = 1, axis2 = 2)
I2_uni = 0.5*(I1_uni**2 - np.trace(C_uni**2, axis1 = 1, axis2 = 2))
I4v_uni = np.einsum('pij,jk->p', C_uni, V0)
I4w_uni = np.einsum('pij,jk->p', C_uni, W0)

I1_equi = np.trace(C_equi,axis1 = 1, axis2 = 2)
I2_equi = 0.5*(I1_equi**2 - np.trace(C_equi**2, axis1 = 1, axis2 = 2))
I4v_equi = np.einsum('pij,jk->p', C_equi, V0)
I4w_equi = np.einsum('pij,jk->p', C_equi, W0)

I1_str = np.trace(C_str,axis1 = 1, axis2 = 2)
I2_str = 0.5*(I1_str**2 - np.trace(C_str**2, axis1 = 1, axis2 = 2))
I4v_str = np.einsum('pij,jk->p', C_str, V0)
I4w_str = np.einsum('pij,jk->p', C_str, W0)

I1_off = np.trace(C_off,axis1 = 1, axis2 = 2)
I2_off = 0.5*(I1_off**2 - np.trace(C_off**2, axis1 = 1, axis2 = 2))
I4v_off = np.einsum('pij,jk->p', C_off, V0)
I4w_off = np.einsum('pij,jk->p', C_off, W0)
```

## New Loading Protocols
#### Redefined Off-x and Off-y loadings
Off-x: $\lambda_x = \lambda_y^{1/4}$ 

Off-y: $\lambda_y = \lambda_x^{1/4}$ 

#### "Sub-strip"
Sub-strip-x: $\lambda_x = \lambda, \, \lambda_y = 1/\lambda_x^{1/5}$

```python
#New off-x and off-y
lm1_offx2 = np.linspace(1,1.5,20)
lm2_offx2 = lm1_offx2**(1/4)
lm3_offx2 = 1/(lm1_offx2*lm2_offx2)
lm1_offy2 = lm2_offx2
lm2_offy2 = lm1_offx2
lm3_offy2 = lm3_offx2
lm1_off2 = np.hstack((np.flip(lm1_offx2,axis=0), lm1_offy2))
lm2_off2 = np.hstack((np.flip(lm2_offx2,axis=0), lm2_offy2))
lm3_off2 = np.hstack((np.flip(lm3_offx2,axis=0), lm3_offy2))

#New sub-strip-x and sub-strip-y
lm1_ssx = np.linspace(1,1.5,20)
lm2_ssx = lm1_ssx**(-1/5)
lm3_ssx = 1/(lm1_ssx*lm2_ssx)
lm1_ssy = lm2_ssx
lm2_ssy = lm1_ssx
lm3_ssy = lm3_ssx
lm1_ss = np.hstack((np.flip(lm1_ssx,axis=0), lm1_ssy))
lm2_ss = np.hstack((np.flip(lm2_ssx,axis=0), lm2_ssy))
lm3_ss = np.hstack((np.flip(lm3_ssx,axis=0), lm3_ssy))

F_off2, C_off2 = matrix_from_stretch(lm1_off2, lm2_off2, lm3_off2)
F_ss, C_ss = matrix_from_stretch(lm1_ss, lm2_ss, lm3_ss)

I1_off2 = np.trace(C_off2,axis1 = 1, axis2 = 2)
I2_off2 = 0.5*(I1_off2**2 - np.trace(C_off2**2, axis1 = 1, axis2 = 2))
I4v_off2 = np.einsum('pij,jk->p', C_off2, V0)
I4w_off2 = np.einsum('pij,jk->p', C_off2, W0)

I1_ss = np.trace(C_ss,axis1 = 1, axis2 = 2)
I2_ss = 0.5*(I1_ss**2 - np.trace(C_ss**2, axis1 = 1, axis2 = 2))
I4v_ss = np.einsum('pij,jk->p', C_ss, V0)
I4w_ss = np.einsum('pij,jk->p', C_ss, W0)
```
