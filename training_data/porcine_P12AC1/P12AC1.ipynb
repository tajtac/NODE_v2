# GOH
Gasser et al 2006(https://doi.org/10.1098/rsif.2005.0073)


Incompressible material assumption
$$\Psi_1(J) = p(1-J)$$
$$\Psi_2(\mathbf{C}) = \frac{\mu}{2}(I_1 -3) +\frac{k_1}{2k_2}[e^{(k_2[\kappa I_1 + (1-3\kappa)I_4-1]^2)}-1]$$
Here $I_4 = \mathbf{C}:\mathbf{M} =\mathbf{C}:  \mathbf{a}_0\otimes \mathbf{a}_0$

Let $\mathbf{H} =\kappa \mathbf{I} + (1-3\kappa)\mathbf{M}$ 

and $H = \mathbf{H}:\mathbf{C}  =\kappa I_1 + (1-3\kappa)I_4$ and $E = [{H}-1]$

$$\mathbf{S}_1  = 2\frac{\partial \Psi_{1}}{\partial \mathbf{C}} = -pC^{-1} $$
$$\mathbf{S}_2  =  2\frac{\partial \Psi_{2}}{\partial \mathbf{C}} = \mu \mathbf{I} + 2k_1 e^{k_2{E}^2}[\kappa \mathbf{I}+(1-3\kappa)\mathbf{M}]$$

Total PK2 stress

$$\mathbf{S} = -p\mathbf{C}^{-1} +  \mu\mathbf{I} + 2k_1 e^{k_2{E}^2}{E}[\kappa \mathbf{I}+(1-3\kappa)\mathbf{M}]$$

Total PK1 stress

$$\mathbf{P} = \mathbf{F}\mathbf{S}$$

```python
def GOH_P(par,lam1,lam2):
    mu = par[0]
    k1 = par[1]
    k2 = par[2]
    kappa = par[3]
    alpha = par[4]
    
    #Structure tensor
    a0 = np.array([np.cos(alpha),np.sin(alpha),0])
    M = np.outer(a0,a0)
    
    #Kinematics
    lam3 = 1/(lam1*lam2)
    F = np.array([[(lam1) ,0., 0],[0.,(lam2), 0] ,[0., 0,(lam3)]])
    C = F.T*F
    invF = np.linalg.inv(F)
    invC = np.linalg.inv(C)
    I = np.eye(3)
    
    #Invariants
    I1 = np.trace(C)
    I4 = np.tensordot(C,M)
    
    
    #Evaluate stress
    H = kappa*I1+(1-3*kappa)*I4
    E = H-1
    S2 = mu*I+2*k1*np.exp(k2*E**2)*E*(kappa*I+(1-3*kappa)*M)
    p = S2[2,2]/invC[2,2] #Get pressure  by equating S_33=0
    S = -p*invC+S2

    P =  F*S
    return P
```

```python
def MR_P(par, lm1, lm2):
    C10 = par[0]
    C01 = par[1]
    C20 = par[2]
    lm3 = 1/(lm1*lm2)
    sigma1 = 2*(C10*(lm1**2 - lm3**2) - C01*(1/lm1**2 - 1/lm3**2) +
              2*C20*(lm1**2 - lm3**2)*(lm1**2 + lm2**2 + lm3**2 - 3))
    sigma2 = 2*(C10*(lm2**2 - lm3**2) - C01*(1/lm2**2 - 1/lm3**2) +
              2*C20*(lm2**2 - lm3**2)*(lm1**2 + lm2**2 + lm3**2 - 3))
    sigma = [[(sigma1), 0, 0], [0, (sigma1), 0], [0, 0, 0]]
    F = np.array([[(lm1) ,0., 0],[0.,(lm2), 0] ,[0., 0,(lm3)]])
    FinvT = np.linalg.inv(F).T
    P = sigma*FinvT
    return P
```

```python
def HGO_P(par,lam1,lam2):
    C10 = par[0]
    k1 = par[1]
    k2 = par[2]
    alpha = par[3] #Assume the fibers are orthogonal to each other
    
    #Structure tensor
    v0 = np.array([np.cos(alpha), np.sin(alpha), 0])
    w0 = np.array([np.cos(alpha),-np.sin(alpha), 0])
    V0 = np.outer(v0,v0)
    W0 = np.outer(w0,w0)
    
    #Kinematics
    lam3 = 1/(lam1*lam2)
    F = np.array([[(lam1) ,0., 0],[0.,(lam2), 0] ,[0., 0,(lam3)]])
    C = F.T*F
    invC = np.linalg.inv(C)
    I = np.eye(3)
    
    #Invariants
    I1 = np.trace(C)
    I4 = np.tensordot(C,V0)
    I6 = np.tensordot(C,W0)
    
    Psi1 = C10
    Psi4 = k1*(I4-1)*np.exp(k2*(I4-1)**2)
    Psi6 = k1*(I6-1)*np.exp(k2*(I6-1)**2)
    
    S2 = 2*Psi1*I + 2*Psi4*V0 + 2*Psi6*W0
    p = S2[2,2]/invC[2,2]
    S = S2 - p*invC

    P =  F*S
    return P
```

```python
def Fung_P(par, lam1, lam2):
    lam3 = 1/(lam1*lam2)
    F = np.array([[(lam1) ,0., 0],[0.,(lam2), 0] ,[0., 0,(lam3)]])
    C = F.T*F
    E_11 = 0.5*(C[0,0]-1)
    E_22 = 0.5*(C[1,1]-1)
    E_Z = E_11
    E_theta = E_22
    
    C  = par[0]
    a1 = par[1]
    a2 = par[2]
    a4 = par[3]
    
    Q = a1*E_theta**2 + a2*E_Z**2 + 2*a4*E_theta*E_Z
    S_theta = C*(a1*E_theta + a4*E_Z)*np.exp(Q)
    S_Z     = C*(a4*E_theta + a2*E_Z)*np.exp(Q)
    S = [[(S_theta), 0, 0], [0, (S_Z), 0], [0, 0, 0]]
    P = F*S
    return P
```

```python
class obj_fun():
    def __init__(self, A, model):
        self.lam1  = A[:,0]
        self.lam2  = A[:,2]
        self.PE1   = A[:,1]
        self.PE2   = A[:,3]
        self.model = model
    def loss(self, par):
        lam1  = self.lam1
        lam2  = self.lam2
        PE1   = self.PE1
        PE2   = self.PE2
        model = self.model
        err = 0.0
        for i in range(0, len(lam1)):
            P = model(par, lam1[i],lam2[i])
            err += (P[0,0]-PE1[i])**2+ (P[1,1]-PE2[i])**2
        return err/len(lam1)
parameters =[]
```

```python
def plot_outputs(title, model, mask_trn, mask_val):
    fig,ax = plt.subplots(1,2,figsize=(6,3),sharey=True)

    lam1_1 = A[mask_trn,0]
    lam1_2 = A[mask_trn,2]
    PE1_1 = A[mask_trn,1]
    PE1_2 = A[mask_trn,3]
    Parr1 = np.zeros([len(lam1_1),2])
    for i in range(0,len(lam1_1)):
            P = model(opt.x,lam1_1[i],lam1_2[i])
            Parr1[i,0] = P[0,0]
            Parr1[i,1] = P[1,1]


    lam2_1 = A[mask_val,0]
    lam2_2 = A[mask_val,2]
    PE2_1 = A[mask_val,1]
    PE2_2 = A[mask_val,3]
    Parr2 = np.zeros([len(lam2_1),2])
    for i in range(0,len(lam2_1)):
            P = model(opt.x,lam2_1[i],lam2_2[i])
            Parr2[i,0] = P[0,0]
            Parr2[i,1] = P[1,1]

    ax[0].plot(lam1_1,PE1_1,color='r',label=r'$\mathbf{e}_1$')       
    ax[0].plot(lam1_1,Parr1[:,0],'--',color='r')
    ax[0].plot(lam1_2,PE1_2,color='b',label=r'$\mathbf{e}_2$')       
    ax[0].plot(lam1_2,Parr1[:,1],'--',color='b') 
    ax[0].set_title("Training")
    ax[0].set_xlabel(r'Stretch $\lambda$')
    ax[0].set_ylabel(r'Nominal Stress $P$[MPa]')
    ax[0].legend()

    ax[1].plot(lam2_1,PE2_1,color='r',label=r'$\mathbf{e}_1$')       
    ax[1].plot(lam2_1,Parr2[:,0],'--',color='r')
    ax[1].plot(lam2_2,PE2_2,color='b',label=r'$\mathbf{e}_2$')       
    ax[1].plot(lam2_2,Parr2[:,1],'--',color='b')  
    ax[1].set_title("Validation")
    ax[1].set_xlabel(r'Stretch $\lambda$')
    
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()
```
